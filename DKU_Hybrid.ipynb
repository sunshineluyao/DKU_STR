# Visualization

```
# import packages
import pandas as pd
import numpy as np
```

```
# import data
df = pd.read_csv('https://raw.githubusercontent.com/sunshineluyao/DKU_STR/main/random_data.csv')
df.head()
```

```
df.describe(include='all')
```

### Q1 Which of the following best describes the instruction that you received for this course?	

```
df_Q1 =pd.DataFrame(df['Q1'].value_counts())
df_Q1
```

```
df_Q1.loc[1.0, 'Answers']=1
df_Q1.loc[2.0, 'Answers']=2
df_Q1.loc[3.0, 'Answers']=3
df_Q1.loc[4.0, 'Answers']=4
df_Q1.loc[5.0, 'Answers']=5
df_Q1.rename(columns={'Q1':'Frequency'},inplace=True)
df_Q1
```

```
df_Q1=df_Q1.sort_values(by='Answers')
df_Q1
```

```
import plotly.express as px
fig = px.bar(df_Q1, x="Answers", y="Frequency", color='Answers', title="Q1:Which of the following best describes the instruction that you received for this course?")
fig.show()
```

### Q6: Did this course include the following group activities? (Select all that applies) - Selected Choice

```
df_Q6= pd.DataFrame(df['Q6_1'].value_counts())
df_Q6['Q6_2']=df['Q6_2'].value_counts()
df_Q6['Q6_3']=df['Q6_3'].value_counts()
df_Q6['Q6_4']=df['Q6_4'].value_counts()
df_Q6['Q6_5']=df['Q6_5'].value_counts()
df_Q6['Q6_6']=df['Q6_6'].value_counts()
df_Q6['Answers']=['Yes','No']
df_Q6_Melt=df_Q6.melt(id_vars=['Answers'], value_vars=['Q6_1','Q6_2','Q6_3','Q6_4','Q6_5','Q6_6'])
df_Q6_Melt.columns=['Answers','Group_Activities','Frequency']
df_Q6_Melt
```

```
import plotly.express as px
fig = px.bar(df_Q6_Melt, x="Group_Activities", y="Frequency", color='Answers', title="Q6: Did this course include the following group activities? (Select all that applies) - Selected Choice?",barmode='group')
fig.show()
```
